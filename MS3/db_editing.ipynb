```python
import pandas as pd
import numpy as np
```

```python
cleaned_df = pd.read_csv('../cleaned_fpl.csv')
df = pd.read_csv('../MS2/fpl_two_seasons.csv')
```

```python
df.head()
```

```python
cleaned_df.head()
```

```python
# add the 'Team' column from cleaned_df to df based on the joins on 'player_id'='element', 'season' and 'fixture'
df = df.merge(cleaned_df[['player_id', 'season', 'fixture', 'team']], left_on=['element', 'season', 'fixture'], right_on=['player_id', 'season', 'fixture'], how='left')
df.head()
```

```python
df[df['team'].isnull()].head()
```

```python
df[(df['team']!=df['home_team']) & (df['team']!=df['away_team'])].head()
```

```python
# cumsum the 'total_points', 'goals_scored', 'assists', 'minutes', 'bonus' and 'clean_sheets' column grouped by 'element' and 'season' to create a new columns
df[['total_points_sum', 'goals_scored_sum', 'assists_sum', 'minutes_sum', 'bonus_sum', 'clean_sheets_sum']] = df.groupby(['element', 'season'])[['total_points', 'goals_scored', 'assists', 'minutes', 'bonus', 'clean_sheets']].cumsum()
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['player_id'], inplace=True)
```

```python
# Turn each row into a textual feature that sumarrizes the key statistics of that player up to that fixture for embedding purposes
df['feature_text'] = (
    "Player ID: " + df['element'].astype(str) +
    ", Player Name: " + df['name'].astype(str) +
    ", Season: " + df['season'].astype(str) +
    ", Team: " + df['team'].astype(str) +
    ", Gameweek: " + df['GW'].astype(str) +
    ", Position: " + df['position'].astype(str) +
    ", Fixture: " + df['fixture'].astype(str) +
    ", Home Team: " + df['home_team'].astype(str) +
    ", Away Team: " + df['away_team'].astype(str) +
    ", Kickoff Time: " + df['kickoff_time'].astype(str) +
    ", Total Points This Gameweek: " + df['total_points'].astype(str) +
    ", Goals Scored This Gameweek: " + df['goals_scored'].astype(str) +
    ", Assists This Gameweek: " + df['assists'].astype(str) +
    ", Minutes Played This Gameweek: " + df['minutes'].astype(str) +
    ", Bonus Points This Gameweek: " + df['bonus'].astype(str) +
    ", Clean Sheets This Gameweek: " + df['clean_sheets'].astype(str) +
    ", Yellow Cards This Gameweek: " + df['yellow_cards'].astype(str) +
    ", Red Cards This Gameweek: " + df['red_cards'].astype(str) +
    ", Own Goals This Gameweek: " + df['own_goals'].astype(str) +
    ", Penalties Saved This Gameweek: " + df['penalties_saved'].astype(str) +
    ", Penalties Missed This Gameweek: " + df['penalties_missed'].astype(str) +
    ", Saves: " + df['saves'].astype(str) +
    ", Form Score: " + df['form'].astype(str) +
    ", Value: " + df['value'].astype(str) +
    ", bonus point system: " + df['bps'].astype(str) +    
    ", ict_index: " + df['ict_index'].astype(str) +
    ", Influence: " + df['influence'].astype(str) +
    ", Creativity: " + df['creativity'].astype(str) +
    ", Threat: " + df['threat'].astype(str) +
    ", Total Points So Far: " + df['total_points_sum'].astype(str) +
    ", Goals Scored So Far: " + df['goals_scored_sum'].astype(str) +
    ", Assists So Far: " + df['assists_sum'].astype(str) +
    ", Minutes Played So Far: " + df['minutes_sum'].astype(str) +
    ", Bonus Points So Far: " + df['bonus_sum'].astype(str) +
    ", Clean Sheets So Far: " + df['clean_sheets_sum'].astype(str)
)
```
